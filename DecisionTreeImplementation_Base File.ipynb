```python
#Function to find label for a value
#if MIN_Value <=val < (m + Mean_Value) / 2 then it is assigned label a
#if (m + Mean_Value) <=val < Mean_Value then it is assigned label b
#if (Mean_Value) <=val < (Mean_Value + MAX_Value)/2 then it is assigned label c
#if (Mean_Value + MAX_Value)/2 <=val <= MAX_Value  then it is assigned label d

def label(val, *boundaries):
    if (val < boundaries[0]):
        return 'a'
    elif (val < boundaries[1]):
        return 'b'
    elif (val < boundaries[2]):
        return 'c'
    else:
        return 'd'

#Function to convert a continuous data into labelled data
#There are 4 lables  - a, b, c, d
def toLabel(df, old_feature_name):
    second = df[old_feature_name].mean()
    minimum = df[old_feature_name].min()
    first = (minimum + second)/2
    maximum = df[old_feature_name].max()
    third = (maximum + second)/2
    return df[old_feature_name].apply(label, args= (first, second, third))
```

```python
#Convert all columns to labelled data
df['sl_labeled'] = toLabel(df, 'sl')
df['sw_labeled'] = toLabel(df, 'sw')
df['pl_labeled'] = toLabel(df, 'pl')
df['pw_labeled'] = toLabel(df, 'pw')
df
```

```python
# function to find the gain ratio
# of a particular split using a given feature and the given value in that feature

def FindGainRatio(feature,df,y,prev_gain):
    
    # making Temporary splits in the
    # dataframe to calculate the gain ratio
    # for the feature that will be used for the split
    split1=pd.DataFrame()
    split2=pd.DataFrame()
    split3=pd.DataFrame()
    split4=pd.DataFrame()
    
    # This section of code uses a for loop
    # to split the dataframe in two parts using
    # the provided feature in the function on the 
    # basis of the value namely a,b,c and d
    for iterator in range(len(df)):
        if df.iloc[iterator][feature]=='a':
            split1=split1.append(y.iloc[iterator],ignore_index=True)
        elif df.iloc[iterator][feature]=='b':
            split2=split2.append(y.iloc[iterator],ignore_index=True)
        elif df.iloc[iterator][feature]=='c':
            split3=split3.append(y.iloc[iterator],ignore_index=True)
        elif df.iloc[iterator][feature]=='d':
            split4=split4.append(y.iloc[iterator],ignore_index=True)
    
    # findind the length of each split
    len1=len(split1)
    len2=len(split2)
    len3=len(split3)
    len4=len(split4)
    length=[len1,len2,len3,len4]
    total=len1+len2+len3+len4
    
    # declaring a few variables to calculate the gain ratio
    # split_info carries the split information of the split
    # Newinfo_gain carries the information gain of the temporary split
    split_info=0
    Newinfo_gain=0
    
    # this section of code first calculates the number of classes
    # in each split and then calculates the split_info and Newinfo_gain
    # for every split.
    
    #This section of code is repeated four times for the four splits from a node
    
    #calculating for split1
    a,b,c=0,0,0
    for i in range(length[3]):
        if split4.iloc[i][0]==0:
            a+=1
        elif split4.iloc[i][0]==1:
            b+=1
        else: 
            c+=1
    if a==0:
        a=length[3]
    if b==0:
        b=length[3]
    if c==0:
        c=length[3]
    if(length[3]!=0):
        answer=((-1)*(a/length[3])*(math.log2(a/length[3])))+((-1)*(b/length[3])*(math.log2(b/length[3])))+((-1)*(c/length[3])*(math.log2(c/length[3])))
        split_info+=((-1)*(length[3]/total)*(math.log2(length[3]/total)))
        Newinfo_gain+=((length[3]/total)*answer) 
    
    
    #calculating for split2
    a,b,c=0,0,0
    for i in range(length[0]):
        if split1.iloc[i][0]==0:
            a+=1
        elif split1.iloc[i][0]==1:
            b+=1
        else: 
            c+=1
    if a==0:
        a=length[0]
    if b==0:
        b=length[0]
    if c==0:
        c=length[0]
    if(length[0]!=0):
        answer=((-1)*(a/length[0])*(math.log2(a/length[0])))+((-1)*(b/length[0])*(math.log2(b/length[0])))+((-1)*(c/length[0])*(math.log2(c/length[0])))
        split_info+=((-1)*(length[0]/total)*(math.log2(length[0]/total)))
        Newinfo_gain+=((length[0]/total)*answer)
    
    
    #calculating for split3
    a,b,c=0,0,0
    for i in range(length[1]):
        if split2.iloc[i][0]==0:
            a+=1
        elif split2.iloc[i][0]==1:
            b+=1
        else: 
            c+=1
    if a==0:
        a=length[1]
    if b==0:
        b=length[1]
    if c==0:
        c=length[1]
    if(length[1]!=0):
        answer=((-1)*(a/length[1])*(math.log2(a/length[1])))+((-1)*(b/length[1])*(math.log2(b/length[1])))+((-1)*(c/length[1])*(math.log2(c/length[1])))
        split_info+=((-1)*(length[1]/total)*(math.log2(length[1]/total)))
        Newinfo_gain+=((length[1]/total)*answer)
    
    
    #calculating for split4
    a,b,c=0,0,0
    for i in range(length[2]):
        if split3.iloc[i][0]==0:
            a+=1
        elif split3.iloc[i][0]==1:
            b+=1
        else: 
            c+=1
    if a==0:
        a=length[2]
    if b==0:
        b=length[2]
    if c==0:
        c=length[2]
    if(length[2]!=0):
        answer=((-1)*(a/length[2])*(math.log2(a/length[2])))+((-1)*(b/length[2])*(math.log2(b/length[2])))+((-1)*(c/length[2])*(math.log2(c/length[2])))
        split_info+=((-1)*(length[2]/total)*(math.log2(length[2]/total)))
        Newinfo_gain+=((length[2]/total)*answer)
        
        
    # after all the splits, the gain ratio is calculated by using 
    # the acquired information from the splits
    info_gain=prev_gain-Newinfo_gain
    if(split_info!=0):
        gain_ratio=info_gain/split_info
        return gain_ratio
    else:
        return -1
```

```python
def build_tree(fd, y, unused_features,level):
    #base case
    
    # 1. unused is empty
    # caclculating number of classes
    if len(unused_features)==0:
        a,b,c,d,entropy=0,0,0,0,0
        lenj=len(y)
        for i in range(len(y)):
            if y.iloc[i][0]==0:
                a+=1
            elif y.iloc[i][0]==1:
                b+=1
            else: c+=1
        if a>0 : d+=1
        if b>0 : d+=1
        if c>0 : d+=1
        
        # if only more than one class is peresent then
        # entropy is 1 else the entropy is 0
        if d>1 : entropy=1
        print("Level: ",level)
        print("Number of 0: ",a)
        print("Number of 1: ",b)
        print("Number of 2: ",c)
        print("Current Entropy: ",entropy)
        print("Reached Leaf Node")
        print()
        return
    
    #converting the dataframe to a set
    z=y.apply(frozenset,axis=1)
        
    # 2. y contains only one distinct value
    if len(z)==1:
        return
    
    if len(fd)==0:
        return

    #declaring variable
    max_gain=0
    gain=0
    feature = ""
    
    # looping through every possible feature that is unused
    for f in unused_features:
        possible_values = set(fd[f])
        # loop over possible values : val
        # find subset of fd & y with f == val
        # find number of mistakes in this subset 
        # if we predict the most common y as the output
        # find sum of all these mistakes
        # update best feature so that that particular feature
        # makes least number of mistakes
        
        # counting the number of classes present right now
        a,b,c=0,0,0
        for i in range(len(y)):
            if y.iloc[i][0]==0:
                a+=1
            elif y.iloc[i][0]==1:
                b+=1
            else: c+=1
        lenj=len(y)
        
        # if only one class is present the entropy is 0
        # and no need to split further
        if a==lenj or b==lenj or c==lenj:
            print("Level: ",level)
            print("Number of 0: ",a)
            print("Number of 1: ",b)
            print("Number of 2: ",c)
            print("Current Entropy: ",0.0)
            print("Reached Leaf Node")
            print()
            return
        
        if a==0:
            a=lenj
        if b==0:
            b=lenj
        if c==0:
            c=lenj
        
        # calculating the present information required 
        # which will be subtracted from Newinfo_gain of the desired split 
        # to find the information gain
        answer=((-1)*(a/lenj)*(math.log2(a/lenj)))+((-1)*(b/lenj)*(math.log2(b/lenj)))+((-1)*(c/lenj)*(math.log2(c/lenj)))
        gain=FindGainRatio(f,fd,y,answer)
        
        # finding the maximum gain
        if gain>max_gain:
            max_gain=gain
            feature=f      
    # here you should know the best feature
    # print it out
    if gain==-1:
        return
    
    # calculating number of classes present right now
    a,b,c=0,0,0
    for i in range(len(y)):
            if y.iloc[i][0]==0:
                a+=1
            elif y.iloc[i][0]==1:
                b+=1
            else: c+=1
    
    print("Level: ",level)
    print("Number of 0: ",a)
    print("Number of 1: ",b)
    print("Number of 2: ",c)
    print("Current Entropy: ",answer)
    print("Splitting on feature ",feature," using gain ratio ",max_gain)
    print()
    
    # splitting the dataframes
    y1=pd.DataFrame()
    y2=pd.DataFrame()
    y3=pd.DataFrame()
    y4=pd.DataFrame()
    fd1=pd.DataFrame()
    fd2=pd.DataFrame()
    fd3=pd.DataFrame()
    fd4=pd.DataFrame()
    
    # This section of code uses a for loop
    # to split the dataframe in two parts using
    # the provided feature in the function on the 
    # basis of the value namely a,b,c and d
    for iterator in range(len(fd)):
        if fd.iloc[iterator][feature]=='a':
            y1=y1.append(y.iloc[iterator],ignore_index=True)
            fd1=fd1.append(fd.iloc[iterator],ignore_index=True)
        elif fd.iloc[iterator][feature]=='b':
            y2=y2.append(y.iloc[iterator],ignore_index=True)
            fd2=fd2.append(fd.iloc[iterator],ignore_index=True)
        elif fd.iloc[iterator][feature]=='c':
            y3=y3.append(y.iloc[iterator],ignore_index=True)
            fd3=fd3.append(fd.iloc[iterator],ignore_index=True)
        elif fd.iloc[iterator][feature]=='d':
            y4=y4.append(y.iloc[iterator],ignore_index=True)
            fd4=fd4.append(fd.iloc[iterator],ignore_index=True)
    
    # this section of code of drops the feature which is
    # used for the split in the newly formed dataframes for 
    # further operations
    if len(fd1)!=0:
        fd1.drop(feature,axis=1,inplace=True)
    if len(fd2)!=0:
        fd2.drop(feature,axis=1,inplace=True)
    if len(fd3)!=0:
        fd3.drop(feature,axis=1,inplace=True)
    if len(fd4)!=0:
        fd4.drop(feature,axis=1,inplace=True)
    
    # removing the feature used for split from the 
    # unused_featured=s set
    unused_features.remove(feature)
    
    
    # making recursive calls for further operations
    # on every newly formed dataframes
    build_tree(fd1,y1,unused_features,level+1)
    build_tree(fd2,y2,unused_features,level+1)
    build_tree(fd3,y3,unused_features,level+1)
    build_tree(fd4,y4,unused_features,level+1)
    # remove best feature from unused features
    # loop over possible values of best feature
    # call build tree recursively
```
